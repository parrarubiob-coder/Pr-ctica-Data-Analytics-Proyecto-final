```python
# Analizo los pedidos entregados que no tienen fecha de entrega
# para comprobar si presentan otras anomalías.

orders[
    (orders["order_status"] == "delivered") &
    (orders["order_delivered_customer_date"].isnull())
]
```

```python
# Elimino los pedidos con estado "delivered" que no disponen de fecha
# de entrega. Estos registros representan únicamente 8 observaciones
# (0,008 % del total) y no permiten calcular correctamente las métricas
# relacionadas con los tiempos de entrega.

orders = orders[
    ~(
        (orders["order_status"] == "delivered") &
        (orders["order_delivered_customer_date"].isnull())
    )
]
```

```python
# Conclusión:
# Se eliminan 8 pedidos con estado "delivered" que no disponían de
# fecha de entrega. Dado que representan un porcentaje muy reducido
# del conjunto de datos (0,008 %), su eliminación no afecta a la
# representatividad del análisis y permite calcular correctamente
# las métricas relacionadas con los tiempos de entrega.
```

```python
# Analizo los pedidos que no tienen fecha de aprobación.

orders[orders["order_approved_at"].isnull()]["order_status"].value_counts()
```

```python
# Se detectan 14 pedidos con estado "delivered" que no disponen de
# fecha de aprobación. Dado que todos ellos presentan fechas de envío
# y entrega, se considera que la aprobación del pedido tuvo lugar.
#
# Para evitar la pérdida de estos registros y mantener la coherencia
# temporal del análisis, se sustituye la fecha de aprobación por la
# fecha de compra del pedido. Esta decisión afecta únicamente al
# 0,014 % del conjunto de datos.

# Sustituyo la fecha de aprobación por la fecha de compra
# en los pedidos entregados que no disponen de fecha de aprobación.

mask = (
    (orders["order_status"] == "delivered") &
    (orders["order_approved_at"].isnull())
)

orders.loc[mask, "order_approved_at"] = orders.loc[
    mask, "order_purchase_timestamp"
]
```

# 5. INTEGRACIÓN DE LOS DATASETS
# ==================================================

```python
# Creo una copia de la tabla orders para construir el dataset final
# sin modificar la tabla original.

df_final = orders.copy()

# Compruebo el tamaño inicial del dataset base.

df_final.shape
```

```python
# Uno la información de pedidos con la información de clientes.
# La unión se realiza mediante customer_id, que está presente en ambas tablas.

df_final = df_final.merge(
    customers,
    on="customer_id",
    how="left"
)

# Compruebo el tamaño del dataset después del merge.

df_final.shape
```

```python
# Compruebo si se han generado valores nulos en las columnas de customers
# después de realizar la unión.

df_final[
    [
        "customer_unique_id",
        "customer_zip_code_prefix",
        "customer_city",
        "customer_state"
    ]
].isnull().sum()
```

```python
# Añado la información de los productos incluidos en cada pedido.
# La unión se realiza mediante order_id.

df_final = df_final.merge(
    items,
    on="order_id",
    how="left"
)

# Compruebo el tamaño del dataset después de añadir items.

df_final.shape
```

```python
# Compruebo cuántos pedidos únicos hay después del merge.

df_final["order_id"].nunique()
```

```python
# Conclusión:
# Tras incorporar la tabla items, el número de filas aumenta de
# 99.433 a 113.417 registros. Este incremento es esperado, ya que
# un mismo pedido puede contener varios productos.
#
# El número de pedidos únicos se mantiene en 99.433, lo que confirma
# que la unión se ha realizado correctamente sin perder información.
```

```python
# Añado la información descriptiva de los productos.
# La unión se realiza mediante product_id.

df_final = df_final.merge(
    products,
    on="product_id",
    how="left"
)

# Compruebo el tamaño del dataset tras la unión.

df_final.shape
```

```python
# Compruebo si existen productos sin información asociada
# después de realizar la unión.

df_final["product_category_name"].isnull().sum()
```

```python
# Compruebo si los registros sin categoría también tienen product_id nulo.

df_final[
    df_final["product_category_name"].isnull()
][["order_id", "order_status", "product_id"]].head(20)
```

```python
# Compruebo que los registros sin información de producto
# corresponden a pedidos cancelados o no disponibles.

df_final[
    df_final["product_id"].isnull()
]["order_status"].value_counts()
```

```python
df_final[
    (df_final["product_id"].isnull()) &
    (df_final["order_status"].isin(["created", "invoiced", "shipped"]))
][["order_id", "order_status"]]
```

```python
# Elimino los pedidos con estados "created", "invoiced" o "shipped"
# que no presentan ninguna línea de producto asociada.

df_final = df_final[
    ~(
        df_final["product_id"].isnull() &
        df_final["order_status"].isin(
            ["created", "invoiced", "shipped"]
        )
    )
]
```

```python
df_final[
    df_final["product_id"].isnull()
]["order_status"].value_counts()
```

```python
# Conclusión:
# Se eliminan 8 pedidos (0,008 % del total) con estados
# "created", "invoiced" o "shipped" que no presentan ninguna
# línea de producto asociada.
#
# Al tratarse de registros inconsistentes y representar un
# porcentaje insignificante del conjunto de datos, su eliminación
# no afecta al análisis y mejora la calidad del dataset final.
```
